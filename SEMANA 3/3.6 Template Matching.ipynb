```python
import os
import numpy as np
from skimage.io import imread, imsave
import matplotlib.pyplot as plt
from skimage.feature import match_template      # <--- match_template
```

```python
im = imread('https://drive.google.com/uc?id=1RQrcAXWdLQXRc2nBqMYxnTKqi5LMpvrp')
im_waldo = imread('https://drive.google.com/uc?id=1YvL3juVOvKLX9Fca-Z09FhlsIiqt0feX')
```

### 1. Cargar las imágenes

```python
plt.imshow(im[:,:,:])
```

```python
plt.imshow(im_waldo)
```

```python
plt.imshow(im_waldo[:,:,0], cmap="gray")
```

### 3. Obtengamos el template matching

Recuede que puede usar la función `match_template`. Hint: Si queremos tener la imágen resultante del mismo tamaño de la imagen original, podemos usar `pad_input=True`

```python
plt.figure(figsize=(20,10))
plt.imshow(match_template(im[:,:,0], im_waldo[:,:,0], pad_input=True), cmap='gray')
plt.colorbar()
```

### 4. Obtengamos la posición de mayor valor, y recortemos +/-100 pixeles alrededor. Mostrar la imagen

```python
# result -> arreglo
result = match_template(im[:,:,0], im_waldo[:,:,0], pad_input=True)
# obtener las coordenadas del valor más alto (punto blanco)
center = np.where(result == np.amax(result))
# obtener el punto aumentado algunos píxeles para ver a Waldo
res = im[center[0][0]-100:center[0][0] + 100, center[1][0]-50:center[1][0]+50]
# mostrar res
plt.imshow(res)
```
