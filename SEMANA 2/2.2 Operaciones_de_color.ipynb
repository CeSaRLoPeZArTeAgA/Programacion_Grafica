# Procesar los colores

```python
import numpy as np
from skimage.io import imread, imsave
import matplotlib.pyplot as plt
```

```python
imagen_original = imread('ORIGINAL.jpeg')
plt.imshow(imagen_original)
```

# Composición por partes

## Rojo

```python
# Imagen Roja:
imagen_roja = imread('ROJO.jpeg')
plt.imshow(imagen_roja)
```

```python
# Parte Roja:
parte_roja = imagen_roja[:,:,0]
plt.imshow(parte_roja, cmap='Reds')
```

## Verde

```python
# Imagen Verde:
imagen_verde = imread('VERDE.jpeg')
plt.imshow(imagen_verde)
```

```python
# Parte Verde:
parte_verde = imagen_verde[:,:,1]
plt.imshow(parte_verde, cmap='Greens')
```

## Azul

```python
# Imagen Azul:
imagen_azul = imread('AZUL.jpeg')
plt.imshow(imagen_azul)
```

```python
# Parte Azul:
parte_azul = imagen_azul[:,:,2]
plt.imshow(parte_azul, cmap='Blues')
```

# Unir todo

```python
# Completar la imagen uniendo las partes roja, verde y azul
imagen_unida = np.dstack((parte_roja, parte_verde, parte_azul))
```

```python
plt.imshow(imagen_unida)
```
