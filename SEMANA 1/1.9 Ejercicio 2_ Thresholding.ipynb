```python
im = skimage.data.coins()
plt.imshow(im, cmap='gray')
```

```python
#@title Cambiar valor de t { run: "auto" }
t = 155 #@param {type:"slider", min:0, max:255, step:1}
im_th = im.copy()
im_th[im_th<t] = 0
im_th[im_th>=t] = 255
# binarizar la imagen (2 valores)
plt.figure()
plt.imshow(im_th, cmap='gray')
```
