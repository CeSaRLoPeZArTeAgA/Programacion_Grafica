```python
url = 'negative_image.jpg'
im = imread(url)
im = im[:,:,0] # <--- color Red (Rojo)
# ::-1 -> invertir
# :250 -> tomar los primeros 250 px
# 250: -> tomar los últimos 250 px
plt.imshow(im, cmap='gray')
```
