```python
#revisar
import gdown
import zipfile

# Descargar archivo desde Google Drive
file_id = "1-GfyMBGbet_xVwzwHSkjNcta5skCmE0X"
url = f"https://drive.google.com/uc?id={file_id}"

output = "reto.zip"
gdown.download(url, output, quiet=False)

# Descomprimir
with zipfile.ZipFile("reto.zip", 'r') as zip_ref:
    zip_ref.extractall("reto")
```

```python
import numpy as np
from skimage.io import imread, imsave
import matplotlib.pyplot as plt

image_stack = []
stack_size = 1000
for i in range(stack_size):
  im = imread(f'reto/im_{i}.png', as_gray=True)
  #im = imread('im_{}.png'.format(i), as_gray = True)
  image_stack.append(im)
```

```python
# 1. image_stack a np.array
image_stack = np.array(image_stack)
image_stack.shape
# 2. media de la imagen -> se acerquen a 0
image = image_stack.mean(axis = 0)
# 3. mostrar la imagen
plt.imshow(image, cmap = 'gray')
```
