## Ejercicio 1

Desarrolle un programa que permita hallar el número mayor de 3 números utilizando estructuras condicionales if.

Nota: Puede hacer uso de IA para estos ejercicios; pero debe comparar que la solución sea óptima y efectivamente solucione el caso esperado.

Nota 2: Si hace uso de IA, debe colocar la respuesta que brindó para cada caso.

```python
import random

#generacion de tres numero aleatorios entre 0 - 30
[a, b, c] = [random.uniform(0, 30) for _ in range(3)]

#impresion de los numeros al azar
print(f"a={a:.2f}")
print(f"b={b:.2f}")
print(f"c={c:.2f}")

#inicializacion con el maximo
maximo=a

#compara con b el max, si no es verdadero pasa a la siguiente pregunta
if b>maximo:
  maximo=b

#compara con c el max y aca termina el algoritmo
if c>maximo:
  maximo=c

#impresion de la respuesta
print(f"El numero mayor es {maximo:.2f}")
```

## Ejercicio 2
Desarrolle el ejercicio 1 con listas y funciones de listas.

```python
#generacion de tres numero aleatorios entre 0 - 30
lista_numeros = [random.uniform(0, 30) for _ in range(3)]

#impresion de numeros al azar
print(f"Numero al azar como lista:{lista_numeros}")

#impresion del maximo con funcion
print(f"maximo valor con funcion={max(lista_numeros):.3f}")

#busqueda de maximo como lista
maximo=lista_numeros[0]
for i in range(1,len(lista_numeros)):
  if lista_numeros[i]>maximo:
    maximo=lista_numeros[i]

#impresion del maximo con listas
print(f"maximo valor lista={maximo:.3f}")
```

## Ejercicio 3
Desarrolle el ejercicio 1 sin if y únicamente operaciones aritméticas.

```python
import random

#funcion que calcula el maximo de dos numeros
def maximo2(a,b):
  return (a+b+abs(a-b))/2

#generacion de tres numero aleatorios entre 0 - 30
[a, b, c] = [random.uniform(0, 30) for _ in range(3)]

#impresion de los numeros al azar
print(f"a={a:.2f}")
print(f"b={b:.2f}")
print(f"c={c:.2f}")

#impresion de maximo
print(f"El numero maximo es:{maximo2(maximo2(a,b),c):.2f}")
```

## Ejercicio 4
Elabore un código que devuelva la letra inicial de cada palabra que contenga mayúsculas y con ellas forme una palabra de solo mayúsculas.

Ejemplo

`Universidad Nacional de Ingeniería`

debe convertirse en:

`UNI`

Nota: se recomienda el uso de funciones.

```python
#nombre al cual queremos sacar sus iniciales
nombre="Universidad Nacional de Ingeniería"
print(nombre)

#separacion del nombre en palabras si el espacion en blanco
palabras_separadas = nombre.split(" ")
print(f"El nombre separado sin espacios= {palabras_separadas}")

#lista de palabras a excluir
palabras=["a","en","una","con","una","la","el","de"]
print(f"palabras a excluir: {palabras}")

#logica para capturar la primera letra de las palabras del nombre
ini=""
for i in range(len(palabras_separadas)):
  aux=0
  for j in range(len(palabras)):
    if palabras_separadas[i]==palabras[j]:
      aux=1
  if aux==0:
    ini=ini+palabras_separadas[i][:1]

#impresion final del resultado
print(f"Iniciales del nombre={ini}")
```
